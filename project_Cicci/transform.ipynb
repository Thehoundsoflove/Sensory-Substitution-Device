```python
def image_to_waveform(image_path, minfreq, maxfreq, sampling_rate, samples):
    # Convert the figure to gray scale
    image = Image.open(image_path).convert("L")
    
    row = image.size[0]  # Number of rows in the image (width) - frequency
    col = image.size[1]  # Number of columns in the image (height) - time

    # Calculate frequency range and frequency interval
    frange = maxfreq - minfreq
    finv = frange / row  # Frequency increment per row
    

    # Calculate the time interval between column pixels
    # sampling_rate - sampling rate
    # samples_col - samples per column pixel
    
    # T-duration per column pixel
    
    T = samples_col * sampling_rate
    
    # Initialize signal array
    s = np.zeros(row, dtype=complex)

    # Loop over each pixel in the image
    for j in range(samples_col):
        sum_val = 0
        for i in range(row):
            amplitude = image.getpixel((i, j))

            # Calculate frequency for the pixel
            f = minfreq + finv * i
            w = 2 * math.pi * f  # Angular frequency

            # Sum up the sinusoidal values weighted by amplitude
            s = math.sin(w * (j/samples_col) * T) * amplitude

        s[j] = sum_val  # Store the summed value for this column
```
