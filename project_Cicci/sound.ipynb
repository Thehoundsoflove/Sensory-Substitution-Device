```python
import numpy as np
import matplotlib.pyplot as plt
import math, array
from PIL import Image, ImageOps
import cv2

# Draw circle

fig, ax = plt.subplots(figsize=(2, 2), facecolor='black')
circle = plt.Circle((0.5, 0.5), 0.4, fill=True, edgecolor='white', facecolor='blue', linewidth=5)
ax.add_patch(circle)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal', adjustable='box')
ax.set_facecolor('black')
plt.show()
png_file_path = 'circle.png'
fig.savefig(png_file_path, format='png', facecolor='black', bbox_inches='tight')
plt.close(fig)

bmp_file_path = 'circle.bmp'
image = Image.open(png_file_path)
image.save(bmp_file_path, format='BMP')
# image = Image.open(png_file_path).convert("RGB")
image = Image.open("circle.bmp").convert("L")
image.show()

row = image.size[0]
col = image.size[1]
amplitudes = []

for y in range(col):
    for x in range(row):
        amplitude = image.getpixel((x,y))
        amplitudes.append(amplitude)
```

```python
def wave_generator(w, samples_col, T, amplitude):
    set1 = []
    for j in range(samples_col):
        F = math.sin(float(w * (j/samples_col) * T)) * float(amplitude)
        F = int(math.floor(F))
        set1.append(F)
    
    return set1

row = image.size[0]
col = image.size[1]
amplitudes = []

minfreq = 40    # Hz
maxfreq = 5000  # Hz
frange = maxfreq - minfreq
finv = frange / row

samples_col = 10
sampling_rate = 10000 
T = samples_col * sampling_rate
waves = []
data = []
test = []
for y in range(col):
    # waves = []
    for x in range(row):
        amplitude = image.getpixel((x,y))
        amplitudes.append(amplitude)
        f = minfreq + finv * (x)
        w = 2 * math.pi * f
        if amplitude > 0:
            wave = wave_generator(w, samples_col, T, amplitude)
            print(len(wave))
            waves.append(wave)

    for i in range(samples_col):
        print(i)
        for wave in waves:
            try:
                test.append(i+y*samples_col)
                data[i + y * samples_col] += wave[i]
            except(IndexError):
                data.insert(i + samples_col * y, wave[i])
```
