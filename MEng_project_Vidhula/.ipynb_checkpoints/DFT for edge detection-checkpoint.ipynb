## DFT for edge detection

**Aim:** Edge detection

**Strategy:**
- Convert image to numpy array and perform Fourier Transform to convert to frequency domain. 
- Pass this through a High Pass Filter for edge detection
- Inverse fourier transform to re-draw image after edge detection

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2

img = Image.open('drawing.png').convert("L") #converted to grayscale
```

```python
# #Thresholding
# threshold_value = 128  # Example threshold value
# _, binary_image = cv2.threshold(np.array(img), threshold_value, 255, cv2.THRESH_BINARY)

# Convert the image to a NumPy array
img_array = np.array(img)

# Display the original image
plt.imshow(img, cmap = 'gray')
plt.title('Original Image')
plt.axis('off')
```

### Discrete Fourier Transform on Image

```python
# Perform the Fourier Transform
f_transform = np.fft.fft2(img_array)
# Shift zero frequency to the center (i.e remove the DC component)
f_shifted = np.fft.fftshift(f_transform)
```

### High-pass filter

```python
rows, cols = img_array.shape

# Determine center of the image
origin_x = rows // 2
origin_y = cols // 2

# Create a mask with a small center area set to 0, high frequency areas set to 1
mask = np.ones((rows, cols)) #intialized to all 1s
radius = min(rows, cols) * 0.1 # Radius for the low-frequency region to block = 10% of the smaller dimension
#increase radius for sharper edges but loss in low- frequency components

y, x = np.ogrid[:rows, :cols]
#Each pair (y[i], x[j]) represents the coordinates of pixel (i, j) in the original image.

#Mask that blocks the low-frequency region
mask_area = (x - origin_y)**2 + (y - origin_x)**2 <= radius**2
mask[mask_area] = 0

f_shifted_filtered = f_shifted * mask
```

### Inverse Fourier Transform to get the edge-detected image

```python
# Inverse FFT
f_ishifted = np.fft.ifftshift(f_shifted_filtered)  # Shift back (i.r reintroduce the DC component)
img_back = np.fft.ifft2(f_ishifted)                # Inverse Fourier Transform
img_back = np.abs(img_back)                        # Take magnitude to get real values
img_back = 1 - img_back                            #Invert image for black on white
```

```python
# Plot original and edge-detected images
plt.figure(figsize=(12, 6))

# Display the original image
plt.subplot(1, 2, 1)
plt.imshow(img_array, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Display the edge-detected image
plt.subplot(1, 2, 2)
plt.imshow(img_back, cmap='gray')
plt.title('Edge Detected Image')
plt.axis('off')

plt.show()
```
